# Class 5: ReAct Agents with Memory (OpenAI / LangChain)
 
**Goal:** Learn how ReAct agents in LangChain handle memory — that is, how agents can remember previous parts of a conversation or interaction.
 
**What you'll learn**
 - How memory works in a ReAct agent context
 - How to add and use conversational memory (like remembering previous user questions)
 - What types of memory are supported
 - Example: asking the agent about previous messages
 
*Tip:* Make sure your `OPENAI_API_KEY` is set in your environment or notebook secrets!


```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
from langchain_openai import ChatOpenAI
```

```python
checkpointer = InMemorySaver()

model = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
agent = create_agent(model, checkpointer=checkpointer)
```

```python
import uuid

unique_id = str(uuid.uuid4())
print("Generated UUID:", unique_id)
```

```python
config = {"configurable": {"thread_id": unique_id}}
config
```

```python
agent.invoke(
    {"messages": "i am Sachith"},
    config=config
)
```

```python
agent.invoke({"messages": "who am I?"}, config=config)
```
